#1. Sobre o projeto:
Imagine que o time de dados de uma Instituição Financeira possui um arquivo CSV que contém os dados de todos os clientes cadastrados. Porém, o que eles querem é pegar esse arquivo, colocar em um banco de dados local e a partir dele realizar consultas SQL. Assim será possível analisar dados, obter insights e tomar decisões estratégicas.

##1.1 Importando dados:

```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
engine=create_engine('sqlite:///:memory:')
```

```python
#Código omitido

url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
```

```python
import pandas as pd
```

```python
dados = pd.read_csv(url)
```

##2. Entendendo os dados:


```python
dados.head()
```

```python
#Deixando o DataFrame no formato do Banco de Dados:
dados.to_sql('clientes', engine, index=False)
```

```python
#usando a função inspect pra ver se a tabela clientes foi criada:
#Código omitido

inspector = inspect(engine)
print(inspector.get_table_names())
```

##3.Consultando com SQL:


```python
#o que o time de dados quer é separar as pessoas que estão com a categoria de renda preenchidas como "Empregado",
#ou seja, só as pessoas que estão trabalhando no momento.

query = 'SELECT* From clientes WHERE Categoria_de_renda = "Empregado"'
```

```python
#Para ler uma consulta em SQL criaremos a variável empregados = pd.read_sql(). Dentro do parêntese,
#passamos a consulta e o motor do bando de dados, query, engine. Em seguida, a executamos.
empregados = pd.read_sql(query, engine)
```

```python
empregados.to_sql('empregados', con=engine, index=False)
```

```python
empregados
```

```python
pd.read_sql_table('empregados', engine)#para ler a tabela completa
```

```python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])#lendo uma parte da tabela
```

#4.Atualizando o Banco de Dados

```python
query = 'SELECT * FROM clientes'
```

```python
pd.read_sql(query, engine)
```

```python
from sqlalchemy import text

# Tirando um cliente que cancelou a conta do banco de dados:
query = 'DELETE FROM clientes WHERE ID_Cliente = 5008804'
with engine.connect() as conn:
  conn.execute(text(query))
  conn.commit() # Adicionado para garantir que a exclusão seja persistida
```

```python
pd.read_sql_table('clientes', engine)
```

```python
#Atualizando um cadastro de cliente:
#O cliente 5008808 está com o grau de escolaridade como Ensino médio. Porém, ele fez faculdade e deseja mudar para Ensino superior.
from sqlalchemy import text

query = 'UPDATE clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808'
with engine.connect() as conn:
    conn.execute(text(query))
    conn.commit() # Adicionado para garantir que a atualização seja persistida
```

```python
pd.read_sql_table('clientes', engine)
```

```python
# Verificando se o cliente com ID_Cliente = 5008804 foi removido
query_deleted_client = 'SELECT * FROM clientes WHERE ID_Cliente = 5008804'
result_deleted = pd.read_sql(query_deleted_client, engine)
print("Verificação do cliente deletado (deve estar vazio):\n", result_deleted)

# Verificando a atualização do grau de escolaridade para ID_Cliente = 5008808
query_updated_client = 'SELECT ID_Cliente, Grau_escolaridade FROM clientes WHERE ID_Cliente = 5008808'
result_updated = pd.read_sql(query_updated_client, engine)
print("\nVerificação do cliente atualizado (Grau_escolaridade deve ser 'Ensino superior'):\n", result_updated)
```
